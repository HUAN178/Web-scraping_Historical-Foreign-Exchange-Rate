```python
import requests
res = requests.get('https://rate.bot.com.tw/xrt/quote/l6m/USD?Lang=en-US')
print(res.text)
```

```python
import pandas as pd
table = pd.read_html(res.text)
```

```python
table[0]
```

```python
df = table[0]
df
```

```python
df.columns = df.columns.droplevel()
df
```

```python
df = df.drop('Unnamed: 6_level_1',axis=1)
df = df.drop('Unnamed: 7_level_1',axis=1)
df = df.drop('Unnamed: 8_level_1',axis=1)
df
```

```python
df.columns = ['Quoted Date','Currency','Cash Rate:Buying','Cash Rate:Selling','Spot Rate:Buying','Spot Rate:Selling']
df
```

```python
df['Quoted Date'] = pd.to_datetime(df['Quoted Date'],format='%Y/%m/%d')
df
```

```python
df = df.sort_values(by=['Quoted Date'],ascending=True).reset_index(drop=True)
df
```
